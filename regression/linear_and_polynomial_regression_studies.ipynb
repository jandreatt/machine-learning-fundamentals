### Correlação

```python
#calcula o coeficiente de correlação para qualquer conjunto de dados
import numpy as np
import matplotlib.pyplot as plt

def correlacao(x, y):
    media_x = np.mean(x)
    media_y = np.mean(y)

    resultado_x = x - media_x
    resultado_y = y - media_y

    numerador = np.sum(resultado_x * resultado_y)
    denominador = np.sqrt(np.sum(resultado_x ** 2) * np.sum(resultado_y ** 2))
    r = numerador / denominador
    return r
```

### Regressão

```python
#calcula os coeficientes de regressão linear (b0 e b1) para qualquer conjunto de dados
def regressao(x, y):
    media_x = np.mean(x)
    media_y = np.mean(y)

    resultado_x = x - media_x
    resultado_y = y - media_y

    numerador = np.sum(resultado_x * resultado_y)
    denominador = np.sum(resultado_x ** 2)
    b1 = numerador / denominador
    b0 = media_y - (b1 * media_x)

    y_pred = b0 + (b1 * x)  # Equação da regressão
    return b0, b1, y_pred
```

```python
#sem interatividade

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Função para criar gráfico 3D
def criar_grafico_3d(X, y, b):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    #plota os dados reais como pontos vermelhos no gráfico 3D
    ax.scatter(X[:, 0], X[:, 1], y, color='r', label='Dados Reais')

    x1_range = np.linspace(X[:, 0].min(), X[:, 0].max(), 20)
    x2_range = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
    X1, X2 = np.meshgrid(x1_range, x2_range)

    #calcula os valores previstos de y (preço) com base na regressão
    y_pred_surface = b[0] + b[1] * X1 + b[2] * X2

    #plota a superfície de regressão
    ax.plot_surface(X1, X2, y_pred_surface, color='b', alpha=0.5)

    ax.set_xlabel('Tamanho da Casa (m²)')
    ax.set_ylabel('Número de Quartos')
    ax.set_zlabel('Preço da Casa (R$)')
    ax.set_title('Regressão Múltipla')
    ax.legend()

    plt.show()

# Função de regressão múltipla
def regmultipla(X, y):
    #adiciona a coluna de 1s para b0
    X_ = np.column_stack((np.ones(X.shape[0]), X))
    #faz a matriz trasnposta
    X_t = X_.T
    #np.linalg.inv faz matriz inversa (-1)
    X_t_inv = np.linalg.inv(np.dot(X_t, X_))
    #calculando b - conforme a fórmula
    b = np.dot(X_t_inv, np.dot(X_t, y))
    print("Coeficientes:", b)
    criar_grafico_3d(X, y, b)

regmultipla(X, y)
```

```python
import numpy as np
import plotly.graph_objects as go

#função para gráfico 3D interativo
def criar_grafico_3d_interativo(X, y, b, r1, r2):
    # Pontos reais
    scatter = go.Scatter3d(
        x=X[:, 0], #tamanho da casa
        y=X[:, 1], #nº de quartos
        z=y,
        mode='markers',
        marker=dict(size=5, color='red'),
        name='Dados Reais'
    )

    #superfície de regressão
    x1_range = np.linspace(X[:, 0].min(), X[:, 0].max(), 20)
    x2_range = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
    X1, X2 = np.meshgrid(x1_range, x2_range)
    #cálculo - seguindo a fórmula
    Y_pred = b[0] + b[1] * X1 + b[2] * X2

    surface = go.Surface(
        x=X1,
        y=X2,
        z=Y_pred,
        opacity=0.6,
        colorscale='Blues',
        name='Plano de Regressão'
    )

    #apresentando coeficientes de correlação
    texto = (
        f"<b>Coef. de Correlação</b><br>"
        f"Tamanho x Preço: {r1:.4f}<br>"
        f"Quartos x Preço: {r2:.4f}"
    )

    #gráfico
    fig = go.Figure(data=[scatter, surface])
    fig.update_layout(
        title='Regressão Múltipla Interativa',
        scene=dict(
            xaxis_title='Tamanho da Casa (m²)',
            yaxis_title='Número de Quartos',
            zaxis_title='Preço da Casa (R$)'
        ),
        annotations=[
            dict(
                showarrow=False,
                text=texto,
                xref="paper", yref="paper",
                x=0.02, y=0.98,
                align="left",
                bordercolor="black",
                borderwidth=1,
                bgcolor="hotpink",
                opacity=0.8
            )
        ]
    )
    fig.show()

# Função de regressão múltipla
def regmultipla(X, y, r1, r2):
    #adiciona a coluna de 1s para b0
    X_ = np.column_stack((np.ones(X.shape[0]), X))
    #faz a matriz transposta
    X_t = X_.T
    #np.linalg.inv faz matriz inversa (-1)
    X_t_inv = np.linalg.inv(np.dot(X_t, X_))
    #calculando b - conforme a fórmula
    b = np.dot(X_t_inv, np.dot(X_t, y))

    r1 = correlacao(X[:, 0], y)
    r2 = correlacao(X[:, 1], y)

    print("Coeficientes:", b)

    criar_grafico_3d_interativo(X, y, b, r1, r2)

    return b

#Executar
#regmultipla(X, y)

#regressão múltipla usando as correlações já calculadas
b = regmultipla(X, y, r_tamanho_preco, r_quartos_preco)

#para tamanho = 1650 e qtde quartos = 3
tamanho = 1650
quartos = 3

#calculando o preço
preco_estimado = b[0] + b[1] * tamanho + b[2] * quartos
print(f"Preço estimado: R$ {preco_estimado:.4f}")
```

k) Calcule o coeficiente de determinação ou R2 para os dados de treino e teste (veja a função r2_score da biblioteca sklearn.metrics). O que se pode concluir com os resultados?

```python
from sklearn.metrics import r2_score

#previsões para os dados de treino
x_treino = df_treino['x']
x_teste = df_teste['x']

#N=1
b1, b0 = linha_reg1
y_reg1_treino = b1 * x_treino + b0
y_reg1_teste = b1 * x_teste + b0

#N=2
b2, b1, b0 = linha_reg2
y_reg2_treino = b2 * x_treino**2 + b1 * x_treino + b0
y_reg2_teste = b2 * x_teste**2 + b1 * x_teste + b0

#N=3
b3, b2, b1, b0 = linha_reg3
y_reg3_treino = b3 * x_treino**3 + b2 * x_treino**2 + b1 * x_treino + b0
y_reg3_teste = b3 * x_teste**3 + b2 * x_teste**2 + b1 * x_teste + b0

#N=8
b8, b7, b6, b5, b4, b3, b2, b1, b0 = linha_reg8
y_reg8_treino = (
    b8 * x_treino**8 + b7 * x_treino**7 + b6 * x_treino**6 + b5 * x_treino**5 +
    b4 * x_treino**4 + b3 * x_treino**3 + b2 * x_treino**2 + b1 * x_treino + b0
)
y_reg8_teste = (
    b8 * x_teste**8 + b7 * x_teste**7 + b6 * x_teste**6 + b5 * x_teste**5 +
    b4 * x_teste**4 + b3 * x_teste**3 + b2 * x_teste**2 + b1 * x_teste + b0
)

#calculando R²
r2_1_treino = r2_score(df_treino['y'], y_reg1_treino)
r2_1_teste = r2_score(df_teste['y'], y_reg1_teste)

r2_2_treino = r2_score(df_treino['y'], y_reg2_treino)
r2_2_teste = r2_score(df_teste['y'], y_reg2_teste)

r2_3_treino = r2_score(df_treino['y'], y_reg3_treino)
r2_3_teste = r2_score(df_teste['y'], y_reg3_teste)

r2_8_treino = r2_score(df_treino['y'], y_reg8_treino)
r2_8_teste = r2_score(df_teste['y'], y_reg8_teste)

#exibindo os resultados
print("Coeficiente de Determinação (R²):")
print(f"N=1 -> Treino: {r2_1_treino:.4f} | Teste: {r2_1_teste:.4f}")
print(f"N=2 -> Treino: {r2_2_treino:.4f} | Teste: {r2_2_teste:.4f}")
print(f"N=3 -> Treino: {r2_3_treino:.4f} | Teste: {r2_3_teste:.4f}")
print(f"N=8 -> Treino: {r2_8_treino:.4f} | Teste: {r2_8_teste:.4f}")
```

O R2 é uma medida estatística de quão próximos os dados estão da linha de regressão ajustada. Se o R2 de treino for alto e o de teste for muito mais baixo, isso indica que o modelo está se ajustando demais aos dados de treino — ou seja, overfitting. O ideal é ter R² altos em ambos, mas próximos entre si.

Modelos com N mais alto (como N=8) tendem a ter overfitting, especialmente com poucos dados de teste.
